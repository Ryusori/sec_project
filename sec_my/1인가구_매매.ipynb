# 2.매매 주택종합, 아파트

```python
meme   = pd.read_csv("/content/drive/MyDrive/과제3/주택가격지수(매매)_20230110143524.csv", sep = ",")
```

```python
# drop(['칼럼'])
meme = meme.drop(['자치구별(1)'], axis=1)
meme.info()
```

```python
# query("==''")
meme14=meme.query("시점 == 2014")

# reset_index()
meme14=meme14.reset_index()
meme14=meme14.iloc[:,2:]
meme14=meme14[1:]

# rename
meme14.rename(columns={"자치구별(2)":"구별",'종합':'종합_14','아파트':'아파트_14'}, inplace=True)
```

```python
# query("==''")
meme15= meme.query("시점 == 2015")

# reset_index()
meme15= meme15.reset_index()
meme15= meme15.iloc[:,2:]
meme15= meme15[1:]

# rename
meme15.rename(columns={"자치구별(2)":"구별",'종합':'종합_15','아파트':'아파트_15'}, inplace=True)
```

```python
import pandas as pd

# merge( 칼럼1,칼럼2 ,on='')
meme_com=pd.merge(meme14,meme15,on ='구별')
meme_com
```

### { 종합 주택, 아파트 최종 }

```python
# 칼럼의 평균구하기
meme_com["종합"]  = (meme_com["종합_14"]+meme_com["종합_15"])/2
meme_com["아파트"]= (meme_com["아파트_14"]+meme_com["아파트_15"])/2

# 칼럼 뽑기
meme_com = meme_com[["구별",'종합',"아파트"]]

# 로그, 등급화화
import numpy as np
meme_com["종합"] = np.log1p(meme_com["종합"])
labels = [ 5, 4, 3, 2, 1]
bins = pd.cut(meme_com["종합"], 5, labels = labels)
meme_com["종합"] = bins

meme_com["아파트"] = np.log1p(meme_com["아파트"])
labels = [ 5, 4, 3, 2, 1]
bins = pd.cut(meme_com["아파트"], 5, labels = labels)
meme_com["아파트"] = bins
meme_com

# 저장하기
# meme_com.to_csv("서울주택매매현황.csv", index = False)
```

# 3.아파트 단지 수

```python
import pandas as pd
teak = pd.read_csv("/content/drive/MyDrive/과제3/공동주택+현황(층별)(2016년+이전)_20230110142848.csv", sep = ",")
```

```python
teak=teak.iloc[:,1:16] 
teak=teak[4:]
teak=teak[["자치구별(2)","2014","2015"]]

# teak=teak[2:]
teak=teak.reset_index().iloc[:,1:]
teak
```

```python
# 타입 바꾸기기
teak["2014"] = teak["2014"].astype("int")
teak["2015"] = teak["2015"].astype("int")

# 평균 내기기
teak["단지수"]=(teak["2014"]+teak["2015"])/2
teak=teak.drop(["2014","2015"], axis=1)

# rename 하기기
teak.rename(columns={"자치구별(2)":"구별"}, inplace=True)

# 로그화, 등급 나누기
import numpy as np

teak["단지수"] = np.log1p(teak["단지수"])
labels = [ 5, 4, 3, 2, 1]
bins = pd.cut(teak["단지수"], 5, labels = labels)
teak["단지수"] = bins
teak

#저장하기기
# teak.to_csv("아파트단지수.csv", index = False)
```
