#### if 년도 크거나 같거나 작다

```python
def f(row):
    if   row['가입년'] < row['년도']:
        val = 2
    elif row['가입년'] == row['년도']:
        val = 1     
    else: 
        val = 0
    return val   
```

#### if 가입년도 크면 그대로, 0이여도 그대로

```python
def f(rows):
    if  rows['년도확인'] == 2:
        val = 2
    elif rows['년도확인'] == 1 and rows['가입월'] <= rows['월']:
        val =1
    elif  rows['년도확인'] == 1 and rows['가입월'] > rows['월']:
        val = 0  
    else:
        val = 0
    return val
```

```python
mf_t['월확인'] = mf_t.apply(f, axis=1)
mf_t
```
