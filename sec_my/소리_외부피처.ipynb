# 2.매매 주택종합, 아파트

```python
meme   = pd.read_csv("/content/drive/MyDrive/과제3/주택가격지수(매매)_20230110143524.csv", sep = ",")
```

#### drop(['칼럼'])

```python
meme = meme.drop(['자치구별(1)'], axis=1)
```

```python
meme.info()
```

#### query("==''")

```python
meme14=meme.query("시점 == 2014")
meme14=meme14.reset_index()
meme14=meme14.iloc[:,2:]
meme14=meme14[1:]
meme14.rename(columns={"자치구별(2)":"구별",'종합':'종합_14','아파트':'아파트_14'}, inplace=True)
meme14
```

```python
meme15= meme.query("시점 == 2015")
meme15= meme15.reset_index()
meme15= meme15.iloc[:,2:]
meme15= meme15[1:]
meme15.rename(columns={"자치구별(2)":"구별",'종합':'종합_15','아파트':'아파트_15'}, inplace=True)
meme15
```

#### merge( 칼럼1,칼럼2 ,on='')

```python
import pandas as pd
meme_com=pd.merge(meme14,meme15,on ='구별')
meme_com
```

#### 칼럼의 평균구하기

```python
meme_com["종합"]=(meme_com["종합_14"]+meme_com["종합_15"])/2
meme_com["아파트"]=(meme_com["아파트_14"]+meme_com["아파트_15"])/2
```

### 칼럼 뽑기

```python
meme_com = meme_com[["구별",'종합',"아파트"]]
meme_com
```

```python
import numpy as np

meme_com["종합"] = np.log1p(meme_com["종합"])
labels = [ 5, 4, 3, 2, 1]
bins = pd.cut(meme_com["종합"], 5, labels = labels)
meme_com["종합"] = bins
meme_com
```

### { 종합 주택, 아파트 최종 }

```python
import numpy as np

meme_com["아파트"] = np.log1p(meme_com["아파트"])
labels = [ 5, 4, 3, 2, 1]
bins = pd.cut(meme_com["아파트"], 5, labels = labels)
meme_com["아파트"] = bins
meme_com
```

```python
meme_com.to_csv("서울주택매매현황.csv", index = False)
```
